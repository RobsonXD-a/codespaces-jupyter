Considere a função $$ y(x) = x^4 - 4x^3 + x^2 + 1$$ no intervalo [-2,4].

Determine dois intervalos, de amplitude no máximo igual a 1, que contenham (cada um) uma  das 4 raízes da função.

```python
import numpy as np
import matplotlib.pyplot as plt     
%matplotlib inline

x = np.arange(-2,4, 0.1)
y = x**4 - 4*x**3 + x**2 + 1
plt.plot(x,y)
plt.grid()
plt.xlabel("eixo x")
plt.ylabel("eixo y")
```

Olhando o gráfico podemos deduzir que a função cruza a linha $y(x)=0$ nos intervalos $[0,1]$ (no ponto aproximadamente de $x=0.9$), e no intervalo $[3,4]$ (no ponto aproximadamente $x=3.6$).

Considere as funções  $$y1(x) = e^x$$ e $$y2(x) = 3x^2$$ no intervalo  [−2,2].

(a) Qual é o número de raízes da equação $𝑦1(𝑥)=𝑦2(𝑥)$ no intervalo [–2, 2], ou seja, o número de vezes que $𝑦1(𝑥)$ e $𝑦2(𝑥)$ se cruzam no intervalo [–2, 2].

(b) Escreva em qual (quais) intervalos $y2(x)$ é maior ou igual à $y1(x)$, e em qual (quais) intervalos $y2(x)$ é menor ou igual à $y1(x)$.

```python
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline

x = np.arange(-2, 2.1, 0.1)
y1 = np.exp(x)
y2 = 3*x**2
plt.plot(x, y1, 'r')
plt.plot(x, y2, 'b')
plt.grid()
plt.xlabel("eixo x")
plt.ylabel("eixo y")
```

(a) 2 vezes
(b) o intervalo em em que o y2 é maior que y1 é [-0.5, 1.75]
e os intervalos em que o y2 é menor que y1 é [-2, -0.5] e [1.75, 2]

# 3

Com o objetivo de aumentar a produção de alimentos em certa região, uma secretaria de agricultura encomendou a uma equipe de agrônomos um estudo sobre as potencialidades do solo dessa região. Na análise da temperatura do solo, a equipe efetuou medições diárias, **durante quatro dias consecutivos**, em intervalos de uma hora.

**As medições tiveram início às 6 horas da manhã do primeiro dia**. Os estudos indicaram que a temperatura $y$ (medida em graus Celsius), e o tempo $x$ (representando o número de horas decorridas após o início das observações), relacionavam-se através da expressão:

$$y(x)= 26 + 5 \cos(\frac{\pi}{12}  x + \frac{4 \pi}{3}).$$

Desenhe o gráfico acima, e com base nesse gráfico, identifique as afirmativas corretas:

(a) A temperatura do solo, às 6 horas da manhã do primeiro dia, foi de aproximadamente 23,5 ºC.


(b) A função $y(x)$ atinge valor máximo igual a 30 ºC.

(c) A temperatura do solo atingiu o valor máximo, no primeiro dia, às 14 h.

(d) A função $y(x)$ é crescente no intervalo [0,8].

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import math 

x = np.arange(0, 10, 0.1)
y = 26 + 5 * np.cos((math.pi/12)*x + 4*math.pi/3) 
plt.plot(x,y)
plt.grid()
plt.xlabel("tempo")
plt.ylabel("temperatura y") 
```
