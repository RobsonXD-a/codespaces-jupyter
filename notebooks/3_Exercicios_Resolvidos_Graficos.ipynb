#1

Considere a função $$ y(x) = x^4 - 4x^3 + x^2 + 1$$ no intervalo [-2,4].


Determine dois intervalos, de amplitude no máximo igual a 1, que contenham (cada um) uma  das 4 raízes da função.

**RESPOSTA**

Vamos desenhar o gráfico da função acima no intervalo de -2 até 4, incluindo a grade (para uma melhor visualização).

```python
import matplotlib.pyplot as plt                   #importamos a biblioteca matplotlib.pyplot para graficos
%matplotlib inline
import numpy as np                                # importamos a biblioteca numpy para lidar com lista de numeros
x = np.arange(-2,4.1,0.1)                         # x tem início em -2, fim em 4, incremento de 0.1
y = x**4 - 4* x**3 + x**2 +1                      # funcao y (como definida acima)
plt.plot(x,y)                                     # desenhamos a funcao y
plt.grid()                                        # colocamos a grade
plt.xlabel("eixo x")                              # colocamos a legenda no eixo x
plt.ylabel("eixo y")                              # colocamos a legenda no eixo y
```

Olhando o gráfico podemos deduzir que a função cruza a linha $y(x)=0$ nos intervalos $[0,1]$ (no ponto aproximadamente de $x=0.9$), e no intervalo $[3,4]$ (no ponto aproximadamente $x=3.6$).

# 2

Considere as funções  $$y1(x) = e^x$$ e $$y2(x) = 3x^2$$ no intervalo  [−2,2].

(a) Qual é o número de raízes da equação $𝑦1(𝑥)=𝑦2(𝑥)$ no intervalo [–2, 2], ou seja, o número de vezes que $𝑦1(𝑥)$ e $𝑦2(𝑥)$ se cruzam no intervalo [–2, 2].

(b) Escreva em qual (quais) intervalos $y2(x)$ é maior ou igual à $y1(x)$, e em qual (quais) intervalos $y2(x)$ é menor ou igual à $y1(x)$.

**RESPOSTA**

```python
import matplotlib.pyplot as plt                   #importamos a biblioteca matplotlib.pyplot para graficos
%matplotlib inline
import numpy as np                                # importamos a biblioteca numpy para lidar com lista de numeros
x = np.arange(-2,2.1,0.1)                         # x tem início em -2, fim em 2
y1 = np.exp(x)                                    # funcao y1
y2 = 3*(x**2)                                     # funcao y2
plt.plot(x,y1,'r')                                # desenhamos o grafico y1 com cor vermelha
plt.plot(x,y2,'b')                                # desenhamos o grafico y2 com cor azul
plt.grid()                                        # colocamos a grade
plt.xlabel("eixo x")                              # colocamos a legenda no eixo x
plt.ylabel("eixo y")                              # colocamos a legenda no eixo y
```

Olhando o gráfico acima podemos deduzir que:

(a) Resposta para item (a) é 2.

(b)

* no intervalo de -2 até  aproximadamente -0,5, a função $y2$ (função azul) é maior ou igual à função $y1$ (função vermelha);

* no intervalo de aproximadamente -0.5 até aproximadamente 0.9, a função $y2$ é menor ou igual à função $y1$;

* no intervalo de aproximadamente 0.9 até 2, a função $y2$ é maior ou igual à função $y1$.


# 3

Com o objetivo de aumentar a produção de alimentos em certa região, uma secretaria de agricultura encomendou a uma equipe de agrônomos um estudo sobre as potencialidades do solo dessa região. Na análise da temperatura do solo, a equipe efetuou medições diárias, **durante quatro dias consecutivos**, em intervalos de uma hora.

**As medições tiveram início às 6 horas da manhã do primeiro dia**. Os estudos indicaram que a temperatura $y$ (medida em graus Celsius), e o tempo $x$ (representando o número de horas decorridas após o início das observações), relacionavam-se através da expressão:

$$y(x)= 26 + 5 \cos(\frac{\pi}{12}  x + \frac{4 \pi}{3}).$$

Desenhe o gráfico acima, e com base nesse gráfico, identifique as afirmativas corretas:

(a) A temperatura do solo, às 6 horas da manhã do primeiro dia, foi de aproximadamente 23,5 ºC.


(b) A função $y(x)$ atinge valor máximo igual a 30 ºC.

(c) A temperatura do solo atingiu o valor máximo, no primeiro dia, às 14 h.

(d) A função $y(x)$ é crescente no intervalo [0,8].


**RESPOSTA**

Qual intervalo de tempo (valores no eixo $x$) devemos considerar? Como a equipe efetuou medições diárias, **durante quatro dias consecutivos, em intervalos de uma hora**, temos que o valor de $x$ deve variar de 0 até $(4 * 24)$ = 96.

Vamos desenhar o gráfico da função acima nesse intervalo, de 0 até 96, incluindo a grade (para uma melhor visualização).

```python
import matplotlib.pyplot as plt                     #importamos a biblioteca matplotlib.pyplot para graficos
%matplotlib inline
import numpy as np                                  # importamos a biblioteca numpy para lidar com lista de numeros
import math                                         # importamos a biblioteca math para poder usar math.pi (ou seja, para numero Pi)
x = np.arange(0,96.1,0.1)                           # x tem início em 0, fim em 96
y = 26 + 5 * np.cos((math.pi/12)*x + 4*math.pi/3)   # funcao y (como definida acima)
plt.plot(x,y)                                       # desenhamos a funcao
plt.grid()                                          # colocamos a grade
plt.xlabel("tempo (x)")                             # colocamos a legenda no eixo x
plt.ylabel("temperatura (y)")                       # colocamos a legenda no eixo y
```

Observando o gráfico acima, podemos concluir que

- A função $y(x)$ atinge máximo 4 vezes, e que esse máximo é valor maior que 30 ºC. **Logo, a afirmativa (b) é FALSA**.

- Vamos olhar a afirmativa (a). **Note que 6 horas da manhã do primeiro dia, quando as medições tiveram início, corresponde ao valor $x=0$ no nosso gráfico** (pois $x$ representa o número de horas decorridas após o início das observações - como definido no exercício).

- Similarmente, na afirmativa (c), 14h de primeiro dia corresponde ao valor $x=14-6=8$ no nosso gráfico.

- Afirmativa (d) pede para analisar se o gráfico é crescente no intervalo [0,8].

**Para descobrir se afirmativas (a), (b) e (c) são verdadeiras ou falsas, devemos olhar com mais precisão o início do gráfico, digamos no intervalo [0,10]**. Vamos repetir o código acima, só que não precisamos incluir as bibliotecas de novo; e neste código, $x$ vai ter início em 0 e fim em 10.



```python
x = np.arange(0,10.1,0.1)                        # x tem início em 0, fim em 10
y = 26 + 5 * np.cos(math.pi/12*x + 4*math.pi/3)
plt.plot(x,y)
plt.grid()
plt.xlabel("tempo (x)")
plt.ylabel("temperatura (y)")
```


Olhando o gráfico acima, podemos afirmar que:
- A temperatura do solo, às 6 horas da manhã do primeiro dia (ou seja, no ponto $x=0$), foi de aproximadamente 23,5 ºC. Logo (a) é verdadeiro
- A temperatura do solo (valor de $y(x)$) atingiu o valor máximo no ponto $x=8$, ou seja, às 6+8 = 14 h do primeiro dia. Logo, (c) é verdadeiro.
- A função $y(x)$  cresce de $x=0$ até $x=8$ (quando $x=8$ atinge máximo), e depois começa a decrescer. Logo, (d) é verdadeiro também.

Observe que  a função 𝑦(𝑥) de fato não atinge valor máximo igual a 30 ºC.  No gráfico acima podemos ver que o valor máximo da função 𝑦(𝑥) é igual a 31 ºC.

Vamos desenhar cada uma das funções (a) - (e) no intervalo $[-2 \pi, 2 \pi]$ e verificar se a figura dada faz ou não a parte das funções que iremos desenhar. Escolhemos este intervalo pois no gráfico dado $x$ varia de $-2 \pi$ até $2 \pi$. Lembre-se que o valor aproximado de $\pi = 3.14$.

Podemos desenhar funções de itens (a)-(c) num gráfico e funções de itens (d) e (e) no outro.

```python
import matplotlib.pyplot as plt                   #importamos a biblioteca matplotlib.pyplot para graficos
%matplotlib inline
import numpy as np                                # importamos a biblioteca numpy para lidar com lista de numeros
import math                                       # importamos a biblioteca math para poder usar math.pi (ou seja, para numero Pi)
x = np.arange(-2*math.pi,2*math.pi + 0.1, 0.1)    # x vai de -2*Pi até 2*Pi
y1 = 2 * np.sin(x/2)                              # funcao do item a
y2 = 2 * np.sin (2*x)                             # funcao do item b
y3 = 1 + np.sin(2*x)                              # funcao do item c
plt.plot(x,y1,'r')                                # desenhamos o grafico y1 com cor vermelha
plt.plot(x,y2,'b')                                # desenhamos o grafico y2 com cor azul
plt.plot(x, y3,'g')                               # desenhamos o grafico y3 com cor verde
plt.grid()                                        # colocamos a grade
plt.xlabel("eixo x")                              # colocamos a legenda no eixo x
plt.ylabel("eixo y")                              # colocamos a legenda no eixo y
```

Olhando o gráfico acima, podemos ver que a função dada é somente parte da função em vermelho, do item (a).

Isso pois só a função vermelha satisfaz que $y(x) = f(x) = 0$ para  $x = -2\pi$    (aproximadamente  -6.28), para $x=0$, e para $x= 2\pi$ (aproximadamente 6.28).

Ainda, só a função vermelha tem mínimo no ponto $x=-\pi$ (aproximadamente -3.14) e  $y= -2$.

E máximo no ponto  $x= \pi$ (aproximadamente 3.14) e $y=2$.

```python
import matplotlib.pyplot as plt                   #importamos a biblioteca matplotlib.pyplot para graficos
%matplotlib inline
import numpy as np                                # importamos a biblioteca numpy para lidar com lista de numeros
import math                                       # importamos a biblioteca math para poder usar math.pi (ou seja, para numero Pi)
x = np.arange(-2*math.pi,2*math.pi + 0.1, 0.1)    # x vai de -2*Pi até 2*Pi
y1 = 2 * np.cos(x/2)                              # funcao do item d
y2 = 2 * np.cos (2*x)                             # funcao do item e
plt.plot(x,y1,'r')                                # desenhamos o grafico y1 com cor vermelha
plt.plot(x,y2,'b')                                # desenhamos o grágico y2 com cor azul
plt.grid()                                        # colocamos a grade
plt.xlabel("eixo x")                              # colocamos a legenda no eixo x
plt.ylabel("eixo y")                              # colocamos a legenda no eixo y
```
