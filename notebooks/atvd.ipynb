# Lista 5

```python
import pandas as pd
import numpy as np

df = pd.read_csv("https://drive.google.com/u/3/uc?id=1pfK9x9A6eo8Lcht_roB6YNmZr7z1XPEX&export=download")
uf=input("Digite o UF").strip().upper()
atr = input("Digite o atributo que deseja saber").strip().upper()

col_val = [ 'LONGVD', 'EDUC', 'RENDA', 'IDHM' ]

if atr in col_val:
    dados_uf = df[df['UF'] == uf]

    if atr not in dados_uf.columns:
        print(f"O atributo {atr} não existe em nossos bancos de dados.")
    
    else:
        if dados_uf[atr].isnull().all():
            print(f"Não há dados válidos para a coluna {atr} na UF {uf}.")
        else:
            media = dados_uf[atr].mean()
            desv = dados_uf[atr].std()
            percentile = np.percentile(dados_uf[atr].dropna(), 25)

            print(f"Média: {media:.2f}")
            print(f"Desvio padrão: {desv:.2f}")
            print(f"25o percentile: {percentile:.2f}")
else:
    print("Coluna inválida. As opções válidas são: LONGVD, EDUC, RENDA, IDHM.")
        

```
