```python
import pandas as pd
import numpy as np

df = pd.read_csv('idh0.csv')

def calculate_statistics(uf_code, column_name):
    filtered_df = df[df['UF'] == uf_code]
    
    if column_name not in filtered_df.columns:
        print(f"Coluna {column_name} não encontrada.")
        return
    
    data = filtered_df[column_name].dropna() 
    
    mean_value = data.mean()
    std_dev = data.std()
    percentile_25 = np.percentile(data, 25)
    
    print(f"Media {mean_value:.2f}, Desvio padrão {std_dev:.2f}, 25o percentile = {percentile_25:.2f}")

uf_code = input("Digite o código da unidade federativa (UF): ").strip().upper()
column_name = input("Digite o nome da coluna (LONGVD, EDUC, RENDA, IDHM): ").strip().upper()

valid_columns = ['LONGVD', 'EDUC', 'RENDA', 'IDHM']
if column_name in valid_columns:
    calculate_statistics(uf_code, column_name)
else:
    print("Nome da coluna inválido.")
```

Considere o arquivo no URL abaixo:

Este aquivo contém uma tabela várias temperaturas em graus Celsius medidas em diferentes períodos (Obs.: as colunas desta tabela estão separadas por ponto e vírgula (;) e os números possuem vírgula (,) como marcador da casa decimal).

Desenvolva um programa que lê a tabela acima e converte as temperaturas de graus Celsius para Kelvin ou Fahrenheit, dependendo da entrada que o usuário fornecer. Se o usuário entrar com a letra K, você deve fazer a conversão para Kelvin. Caso o usuário fornecer a letra F, você deve fazer a conversão para Fahrenheit. Se nenhuma destas letras foram entradas, o seu programa deve imprimir a mensagem de erro: Opção Inválida! (Obs.: lembre-se de fazer a leitura com o comando input()).

Por fim, após a conversão das temperaturas, imprima na tela as X primeiras temperaturas convertidas que aparecem na tabela, em que X é um valor inteiro fornecido pelo usuário.

Obs.: Como o arquivo do URL acima, ‘temp1.csv’, está anexado neste exercício, basta chamar o seguinte comando para fazer leitura do arquivo: df = pd.read_csv('temp1.csv',sep=';',decimal=',') Desta forma, não é necessário utilizar a url disponibilizada acima.

```python
import pandas as pd

def celsius_to_kelvin(celsius):
    return celsius + 273.15

def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

def main():
    # Carregar o arquivo CSV
    df = pd.read_csv('https://drive.google.com/u/3/uc?id=1nuFiziMXVTLuqWInzWXwPArIEZSJY6oX&export=download', sep=';', decimal=',')
    
    # Ler a opção e o número de registros a serem exibidos
    opcao = input().strip().upper()
    try:
        x = int(input().strip())
    except ValueError:
        print("Número inválido de registros.")
        return

    # Validar a opção e converter as temperaturas
    if opcao == 'K':
        df['Kelvin'] = df['Temperatura'].apply(celsius_to_kelvin)
        print(df[['Periodo', 'Temperatura', 'Kelvin']].head(x).to_string(index=False))
    elif opcao == 'F':
        df['Fahrenheit'] = df['Temperatura'].apply(celsius_to_fahrenheit)
        print(df[['Periodo', 'Temperatura', 'Fahrenheit']].head(x).to_string(index=False))
    else:
        print("Opção Inválida!")

# Verificar se o módulo está sendo executado diretamente
if __name__ == "__main__":
    main()
```
