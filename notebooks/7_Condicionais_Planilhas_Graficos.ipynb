# o Comando `if`

Vamos inicialmente relembrar o comando `if` (para mais detalhes, ver Unidade 2).

Vamos ver um exemplo simples.

Suponha que você queira comprar um pacote de farinha e existem duas opções de tamanho:
* pacotes de 3kg
* pacotes de 5kg.

Suponha que você não se importe tanto em comprar 3kg ou 5kg: o que você quer é comprar a **melhor oferta** (ou seja, a opção com melhor custo-benefício).

E o nosso programa inteiro ficaria assim:

````python
preco5 = float(input("Digite o preço do pacote de 5kg: "))
preco3 = float(input("Digite o preço do pacote de 3kg: "))
razao5  = preco5 / 5
razao3 = preco3 / 3
print("O preço por kg do pacote de 5kg é", razao5)
print("O preço por kg do pacote de 3kg é", razao3)
if razao3 < razao5:
  print("A segunda opção é melhor")
````

**Observação:** Note que, após a condição no ``if``, colocamos um ``:`` (dois-pontos)




### Importante!!
  Os comandos a serem executados quando a condição é verdadeira são identificados pelo número de espaços no começo da linha.

  Eles devem estar **deslocados para a direita**.


O uso de espaços no código é chamado de **indentação**.

Na maioria das linguagens, a indentação tem um papel de organização/estético, mas em Python o uso correto de indentação é crucial.

**Colocar espaços extras ou a menos no começo das linhas pode causar muitos problemas!**

Para deslocar o seu código para a direita é melhor usar o ``TAB``, mas você pode usar a tecla de espaço em branco também.

## Gráficos


Agora vamos ver um exemplo com gráficos.

Lembre-se que a fórmula de Bhaskara nos diz que as raízes de
$$ax^2 + bx+ c = 0$$
são
$$\frac{-b-\sqrt{\Delta}}{2a}$$
e
$$\frac{-b+\sqrt{\Delta}}{2a},$$

onde $\Delta = b^2-4ac$.

As raízes são reais apenas quando $\Delta \geq 0$. Se $\Delta=0$, elas têm o mesmo valor.

Suponha que queremos fazer um gráfico da parábola com x variando da raiz menor para a raiz maior.

Começamos com o cálculo acima:

Digite os valores:
* $a = 2$,  $b = 5$, $c = 1$

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

a = int(input("Digite o valor de a: "))
b = int(input("Digite o valor de b: "))
c = int(input("Digite o valor de c: "))

delta = b*b - 4*a*c
print("Delta =", delta)

r1 = (-b - delta**0.5)/(2*a)
r2 = (-b + delta**0.5)/(2*a)
print("Primeira raiz", r1)
print("Segunda raiz", r2)
````

E então fazemos o gráfico com x variando de r1 a r2.



````python
x = np.arange(r1,r2,0.01)
y = a*x*x + b*x + c
plt.plot(x,y)
````

> Quais são os defeitos que você consegue achar em nosso programa?

> Ele sempre funciona?

Em geral, o primeiro defeito que notamos é que

- se $\Delta<0$, não vai funcionar, pois as raízes não são reais

- se $\Delta=0$, as raízes são iguais e não aparece nada no gráfico.

Execute o código para
- $a=1, b=1, c=2$ ($\Delta <0$)
- $a=1, b=2, c=1$ ($\Delta =0$)

e veja o que acontece. Copiamos o código inteiro abaixo:

````python
a = int(input("Digite o valor de a: "))
b = int(input("Digite o valor de b: "))
c = int(input("Digite o valor de c: "))

delta = b*b - 4*a*c
print("Delta =", delta)

r1 = (-b - delta**0.5)/(2*a)
r2 = (-b + delta**0.5)/(2*a)
print("Primeira raiz", r1)
print("Segunda raiz", r2)

x = np.arange(r1,r2,0.01)
y = a*x*x + b*x + c
plt.plot(x,y)
````

> Como consertar o nosso código?

Um primeiro conserto seria fazer o gráfico apenas no caso em que há duas raízes.

Veja como usar o ``if``.

Note como colocamos vários comandos dentro do bloco do ``if``.

````python
a = int(input("Digite o valor de a: "))
b = int(input("Digite o valor de b: "))
c = int(input("Digite o valor de c: "))

delta = b*b - 4*a*c
print("Delta =", delta)

if delta > 0:
  r1 = (-b - delta**0.5)/(2*a)
  r2 = (-b + delta**0.5)/(2*a)
  print("Primeira raiz", r1)
  print("Segunda raiz", r2)
  x = np.arange(r1,r2,0.01)
  y = a*x*x + b*x + c
  plt.plot(x,y)
````

> E agora o código está perfeito?
> Funciona sempre?

Execute-o com $a=-1, b=4, c=1$ e veja o que acontece.
O código está copiado abaixo.

````python
a = int(input("Digite o valor de a: "))
b = int(input("Digite o valor de b: "))
c = int(input("Digite o valor de c: "))

delta = b*b - 4*a*c
print("Delta =", delta)

if delta > 0:
  r1 = (-b - delta**0.5)/(2*a)
  r2 = (-b + delta**0.5)/(2*a)
  print("Primeira raiz", r1)
  print("Segunda raiz", r2)
  x = np.arange(r1,r2,0.01)
  y = a*x*x + b*x + c
  plt.plot(x,y)
````

> O que você acha que deu errado?

Podemos ver que a primeira raiz é maior que a segunda raiz:
```
Primeira raiz 4.23606797749979
Segunda raiz -0.2360679774997898
```

Então o intervalo para o x é vazio (o número menor deve aparecer primeiro no ``arange``).

Temos que garantir que x vai variar da raiz **menor** para a raiz **maior**:

````python
a = int(input("Digite o valor de a: "))
b = int(input("Digite o valor de b: "))
c = int(input("Digite o valor de c: "))

delta = b**2 - 4*a*c
print("Delta =", delta)

if delta > 0:
  r1 = (-b - delta**0.5)/(2*a)
  r2 = (-b + delta**0.5)/(2*a)
  print("Primeira raiz", r1)
  print("Segunda raiz", r2)

  lim1 = r1
  lim2 = r2
  if r1>r2:
    lim1 = r2
    lim2 = r1
  # neste momento lim1 eh o MENOR valor entre valores r1 e r2. E lim2 eh MAIOR valor entre valores r1 e r2.
  x = np.arange(lim1,lim2,0.01)
  y = a*x*x + b*x + c
  plt.plot(x,y)
````

Note que aqui fazemos ``x`` variar de ``lim1`` para ``lim2``.

Inicialmente, colocamos ``lim1=r1`` e ``lim2=r2``.
Então usamos o ``if`` para trocá-los se necessário.

### Importante!

Note que usamos um  ``if r1>r2:`` dentro do bloco para ``if delta > 0:``.

Então para dizer o que está dentro do bloco do ``if r1>r2:``, nós deslocamos os comandos duplamente para a direita.

# Os comandos ``if`` / ``else``

Relembrando, no exemplo para os pacotes de farinha, mostramos uma mensagem apenas no caso em que ``razao3 < razao5``.

Podemos usar o comando ``else`` para lidar com o caso oposto.

````python
preco5 = float(input("Digite o preço do pacote de 5kg: "))
preco3 = float(input("Digite o preço do pacote de 3kg: "))
razao5  = preco5 / 5
razao3 = preco3 / 3
print("O preço por kg do pacote de 5kg é", razao5)
print("O preço por kg do pacote de 3kg é", razao3)
if razao3 < razao5:
  print("A segunda opção é melhor.")
else:
  print("A primeira opção é melhor.")
````

Os comandos dentro do bloco do ``else`` só serão executados se a condição do ``if`` for falsa.

> Mas, ao invés de ``if/else``, dava para usar dois ``if``'s?

Sim. Logicamente não faz diferença:

````python
preco5 = float(input("Digite o preço do pacote de 5kg: "))
preco3 = float(input("Digite o preço do pacote de 3kg: "))
razao5  = preco5 / 5
razao3 = preco3 / 3
print("O preço por kg do pacote de 5kg é", razao5)
print("O preço por kg do pacote de 3kg é", razao3)
if razao3 < razao5:
  print("A segunda opção é melhor.")
if razao3 >= razao5:
  print("A primeira opção é melhor.")
````

Mas, sempre que puder, use o ``else``.

O código fica mais fácil de ler e tem menos chance de cometermos um erro.

# O comando ``elif``

Voltando para o exemplo dos pacotes de farinha.

Agora suponha que você considera que é um **empate** no caso em que a diferença de ``razao3`` e ``razao5`` seja de até 1 centavo.

Vamos usar o comando ``elif``.

````python
preco5 = float(input("Digite o preço do pacote de 5kg: "))
preco3 = float(input("Digite o preço do pacote de 3kg: "))
razao5  = preco5 / 5
razao3 = preco3 / 3
print("O preço por kg do pacote de 5kg é", razao5)
print("O preço por kg do pacote de 3kg é", razao3)

if razao3 < razao5-0.01:
  print("A segunda opção é melhor.")
elif razao5 < razao3-0.01:
  print("A primeira opção é melhor.")
else:
  print("Empate")
````

Agora o programa faz assim:
- Primeiro ele checa se ``razao3 < razao5-0.01``.

  Se é verdade, então ele executa ``print("A segunda opção é melhor.")``.
  Ele não checa a condição do ``elif`` e não executa o ``else``.

- Se ``razao3 < razao5-0.01`` não vale, então ele checa a condição no ``elif``. Ou seja, se ``razao5 < razao3-0.01``
  Se for verdadeiro, ele executa ``print("A primeira opção é melhor.")``

- Se nenhuma condição do ``if`` e do ``elif`` vale, ele executa o que está no ``else`` (se houver ``else``).


Podemos usar vários ``elif``'s!

Suponha que queremos converter uma nota para conceito com as seguintes regras:
* $nota \geq 95$: conceito A
* $80 \leq nota < 95$: conceito B
* $60 \leq nota < 80$: conceito C
* $50 \leq nota < 60$: conceito D
* $nota < 50$: conceito F


````python
nota = int(input("Nota final: "))
if nota>=95:
  conceito = "A"
elif nota >= 80:
  conceito = "B"
elif nota >=60:
  conceito = "C"
elif nota >=50:
  conceito = "D"
else:
  conceito = "F"
print(f"\033[31m{conceito}\033[0m")
````

# Condições compostas

Suponha que, no nosso exemplo dos pacotes de farinha, agora tenhamos 3 tipos de pacotes:
* Pacotes de 1kg
* Pacotes de 3kg
* Pacotes de 5 kg

Agora temos a seguintes possibilidades:
- Se ``razao3 <= razao5`` **E** ``razao3 <= razao1``, então o pacote com 3 kgs é a melhor oferta
- Se ``razao5 <= razao3`` **E** ``razao5 <= razao1``,  então o pacote com 5 kgs é a melhor oferta:
- Se nenhuma das opções acima vale, então a melhor oferta é o pacote com 1 kg.

Então combinamos as condições usando ``and`` como a seguir.

````python
preco5 = float(input("Digite o preço do pacote de 5kg: "))
preco3 = float(input("Digite o preço do pacote de 3kg: "))
preco1 = float(input("Digite o preço do pacote de 1kg: "))
razao5  = preco5 / 5
razao3 = preco3 / 3
razao1 = preco1 / 1
print("O preço por kg do pacote de 5kg é", razao5)
print("O preço por kg do pacote de 3kg é", razao3)
print("O preço por kg do pacote de 1kg é", razao1)
if razao3 <= razao5 and razao3 <= razao1:
  print("Melhor comprar o pacote 3.")
elif razao5 <= razao3 and razao5 <= razao1:
  print("Melhor comprar o pacote 5.")
else:
  print("Melhor comprar o pacote 1.")
````

``condicao1 and condicao2``  vale verdadeiro quando ambas as condições valem ao mesmo tempo:

````python
1 < 2 and 3==3
````

````python
1 < 2 and 3>4
````

Podemos unir condições usando também um ``or``

``condicao1 or condicao2`` vale verdadeiro quando pelo menos uma das condições é verdadeira (inclusive quando ambas são verdadeiras):

````python
1 < 2 or 3==3
````

````python
1 < 2 or 3>4
````

````python
1 > 2 or 3>4
````
