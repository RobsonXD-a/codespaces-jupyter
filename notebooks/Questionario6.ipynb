Esta tabela contém pesos (kg) e alturas (m) de pessoas hipotéticas. Selecione desta tabela todos os indivíduos do sexo masculino e faça a regressão linear da altura contra o peso para estes indivíduos. Use a altura como variável independente e o peso como variável dependente. Responda as seguintes perguntas:

Qual o valor da inclinação da reta da regressão?

Qual o valor do intercepto da reta de regressão?

De acordo com o nosso modelo de regressão, se uma pessoa tem altura 1.8 m, qual o seu peso?

```python
df = pd.read_csv('https://drive.google.com/u/3/uc?id=17rH9fRiOMttCwRLlVNysDhWNbstPoNlI&export=download', sep=';', decimal=',')
df_male = df[df['sexo'] == 'masculino']
# Pontos a e b
(a, b) = np.polyfit(x=df_male["altura"], y=df_male["peso"], deg = 1)
print(f"{a:.2f} {b:.2f}")
print(f'{a*1.8+b:.2f}')

# Gráfico 
plt.plot(df_male['altura'], df_male['peso'], 'o')

#reta
x = np.arange(1, 2.2)
y = a*x+b
plt.plot(x, y, 'r')
```
